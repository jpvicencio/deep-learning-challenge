## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])

# Display the DataFrame to verify the columns have been dropped
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
print(unique_values)
```

```python
#Identify columns with more than 10 unique values:
for column in application_df.columns:
    if application_df[column].nunique() > 10:
        print(application_df[column].value_counts())
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"

print(application_df['APPLICATION_TYPE'].value_counts())
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 157

# Create a list of application types to be replaced
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
print(application_df['CLASSIFICATION'].value_counts())
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# Filter to find classifications with counts greater than 1
# Count the occurrences of each classification
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Filter for classifications with counts greater than 1
filtered_classifications = classification_counts[classification_counts > 1]

# Display the filtered classifications
print(filtered_classifications)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

cutoff_value = 778

# Count the occurrences of each classification
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Create a list of classifications to be replaced
classifications_to_replace = classification_counts[classification_counts <= cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Encode categorical variables
X = pd.get_dummies(application_df.drop('IS_SUCCESSFUL', axis=1), drop_first=True)  # Features
y = application_df['IS_SUCCESSFUL'].values  # Target

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

Model #1

```python
# Define the model - deep neural net
number_input_features = X_train_scaled.shape[1]  # Number of input features
hidden_nodes_layer1 = 80  # Number of nodes in the first hidden layer
hidden_nodes_layer2 = 30  # Number of nodes in the second hidden layer

# Create the Keras Sequential model
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation='relu', input_dim=number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
# Define the number of epochs and batch size
epochs = 100
batch_size = 32

# Fit the model on the training data
history = nn.fit(X_train_scaled, y_train, epochs=epochs, batch_size=batch_size)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

# Define the model - deep neural net
number_input_features = X_train_scaled.shape[1]  # Number of input features
hidden_nodes_layer1 = 84  # Number of nodes in the first hidden layer
hidden_nodes_layer2 = 42  # Number of nodes in the second hidden layer

# Create the Keras Sequential model
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation='relu', input_dim=number_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
# Define the number of epochs and batch size
epochs = 200
batch_size = 64

# Fit the model on the training data
history = nn.fit(X_train_scaled, y_train, epochs=epochs, batch_size=batch_size)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

Model #3

```python
# Define the model - deep neural net
number_input_features = X_train_scaled.shape[1]  # Number of input features
hidden_nodes_layer1 = 84  # Number of nodes in the first hidden layer
hidden_nodes_layer2 = 42  # Number of nodes in the second hidden layer

# Create the Keras Sequential model
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation='relu', input_dim=number_input_features))

# Second hidden layer with adjusted neurons and dropout
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))
nn.add(tf.keras.layers.Dropout(0.5))  # Adding dropout for regularization

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
# Define the number of epochs and batch size
epochs = 200
batch_size = 64

# Fit the model on the training data
history = nn.fit(X_train_scaled, y_train, epochs=epochs, batch_size=batch_size)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
